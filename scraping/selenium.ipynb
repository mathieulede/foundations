```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import re
import pandas as pd
```

```python
BASE_URL = 'http://compendium.ch'
browser = webdriver.Chrome('/Users/mathieurudaz/Desktop/LEDE/chromedriver')
browser.get(BASE_URL + '/comp/fnr/')
time.sleep(2)
```

```python
soup_doc = BeautifulSoup(browser.page_source, "html.parser")
```

```python
companies_doc_list = soup_doc.find_all(class_='hlSubstance')
```

```python
companies = []

for c in companies_doc_list[:10]:
    company_url = BASE_URL + '/comp/fnr/' + re.findall('\d+', c['href'])[0] + '/fr'
    browser.get(company_url)
    time.sleep(1)
    company_soup_doc = BeautifulSoup(browser.page_source, "html.parser")
    companies.append({
        'name': c.string,
        'products': [{
            'name': p.string, 
            'url': p['href']} for p in company_soup_doc.find_all(class_='company_prod_a')]
    })
    
companies
```

```python
df = pd.DataFrame(companies)
```

```python
df_pharma = pd.concat([pd.DataFrame(c) for c in companies])
```

```python
df_pharma['product'] = pd.DataFrame([p['name'] for p in df_pharma['products']])
```

```python
products = []

for c in companies:
    for p in c['products']:
        product_url = BASE_URL + p['url']
        browser.get(product_url)
        time.sleep(2)
        product_soup_doc = BeautifulSoup(browser.page_source, "html.parser")
        products.append({
            'atc_code': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblAtcCode').string,
            'name': p['name'],
            'description': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblAtcDescr').string,
            'composition': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblComposition').string,
            'indication': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblIndication').string,
            'dosage': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblDosage').string,
            'contraindication': product_soup_doc.find(id='ctl00_MainContent_ucProductDetail1_dvProduct_lblContraindication').string
        })

products[0]
```

```python
import pandas as pd
```

```python
df_directors = pd.read_csv('directors-psql.csv', index_col='id')
df_countries = pd.read_csv('countries-psql.csv', index_col='id')
```

```python
df_directors.shape
```

```python
df_countries[df_countries['common_name'].str.contains('Yugosl')]
```

```python
df_biographies = pd.read_csv('biographies_filtered.txt', error_bad_lines=False)
```

```python
df_merged = df_biographies.merge(df_countries, left_on='nationality', right_on='common_name', how='left')
```

```python
df_merged.shape
```

```python
df_merged[df_merged['code'].isnull()]['nationality'].unique()
```

```python
countries = {
    'West Germany':'DE', 
    'UK':'GB', 
    'USA':'US', 
    'Palestine':'PS', 
    'Taiwan':'TW', 
    'China':'CN',
    'Yugoslavia':'MK'
}
```

```python
df_merged.loc[df_merged['code'].isnull(), 'code'] = df_merged[df_merged['code'].isnull()]['nationality']
```

```python
df_merged['code'].replace(countries, inplace=True)
```

```python
df_merged.drop(['common_name', 'nationality'], axis=1)
```

```python
df = df_merged.merge(df_directors, left_on=['first_name', 'last_name'], right_on=['first_name', 'last_name'], how='right')
```

```python
df_directors.shape
```

```python
df[df['code'].isnull()][['last_name', 'first_name']].sort_values(by='last_name').to_csv('directors_incomplete.txt', index=False)
```
